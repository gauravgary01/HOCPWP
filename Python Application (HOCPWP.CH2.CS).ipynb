```python
import random

def bmiCalc():
    x = float(input('Please enter your weight in Kgs(Kilograms):'))
    y = float(input('Please enter your height in m(meters):'))
    z = x/(y*y)
    if (z < 18.5):
        print('Underweight: less than 18.5')
    elif (z == 18.5 or z <= 24.9):
        print('Normal: between 18.5 and 24.9')
    elif (z > 24.9 or z <= 29.9):
        print('Overweight: between 25 and 29.9')
    elif (z > 29.9):
        print('Obese: 30 or greater')
    else:
        print('Are you a human?')
        

def checkPalindrome():
    num = int(input('Enter number to check:'))
    if (list(str(num)) == list(str(num))[::-1]):
        print('Yes,', num, 'is a palindrome number')
    else:
        print('No,', num, 'is not a palindrome number')
        

def convDecToBin():
    dec_num = float(input('Enter decimal number to convert to binary number:'))
    n = int(input('Enter float position upto:'))
    int_prt = int(dec_num)
    dec_prt = dec_num - int_prt
    b1 = ''
    b2 = ''
    
    while (int_prt > 0):
        r1 = int_prt % 2
        b1 = str(r1) + b1
        int_prt = int_prt // 2
    
    for i in range (n):
        int_2 = int(dec_prt * 2)
        b2 = b2 + str(int_2)
        dec_prt = (dec_prt * 2) - int(dec_prt * 2)
    
    bin_num = b1 + "." + b2
    print('Its binary form is:', bin_num)
    

def checkCC():
    a = int(input('Enter you account number:'))
    b = int(input('Enter your balance at begining of month:'))
    c = int(input('Enter total of all items charged by you this month:'))
    d = int(input('Enter total of all credits applied to your account:'))
    e = int(input('Enter your allowed credit limit:'))
    
    b2 = b + c - d
    
    if (b2 > e):
        print("Credit limit exceeded")
    else:
        print("New balance: ", b2, 'and remaining credit balance is:', e - b2)

# Sorter
    
def initializeList(n):
    return [0]*n


def autoEntry(n):
    lst = list(initializeList(n))
    for i in range(n):
        lst[i] = random.randint(11, 100)
    return lst
        

def manualEntry(n):
    lst = list(initializeList(n))
    for i in range(n):
        lst[i] = int(input(f'Enter integer {i + 1}:'))
    return lst


def bubbleSort(lst):
    n = len(lst)
    for i in range (n):
        for j in range(n-i-1):
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]


def displayList(lst):
    if len(lst) < 100:
        print('The sorted list is:', lst)
        
        
def main():
    n = int(input('Enter list length:'))
    while True:
        choice = input('Select automatic or manual entry (a/m), or press q to quit: ')
        if choice == 'a':
            lst_main = autoEntry(n)
            bubbleSort(lst_main)
            displayList(lst_main)
        elif choice == 'm':
            lst_main = manualEntry(n)
            bubbleSort(lst_main)
            displayList(lst_main)
        elif choice == 'q':
            break
        else:
            print('Invalid input. Please enter a/m/q:')


def displayMenu():
    print('Menu')
    print('1. BMI Calculator')
    print('2. Check Customer Credit')
    print('3. Check a number for Palindrome')
    print('4. Convert a Decimal number to Binary system')
    print('5. Initialize a list and sort it')
    print('6. Exit')
    
    option = int(input("Enter an option: "))
    
    if option == 1:
        bmiCalc()
    elif option == 2:
        checkCC()
    elif option == 3:
        checkPalindrome()
    elif option == 4:
        convDecToBin()
    elif option == 5:
        main()
    elif option == 6:
        exit()
    else:
        print("Invalid option. Please try again.")



while True:
    userInput = input("Do you want to run the program? (y/n): ")
    if userInput.lower() == "y":
        displayMenu()
    if userInput.lower() == "n":
        break
```
